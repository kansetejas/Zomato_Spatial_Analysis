```python
from geopy.geocoders import Nominatim
```

```python
rest_loc = pd.DataFrame()
```

```python
rest_loc['Name'] = df['location'].unique()
```

```python
geolocator = Nominatim(user_agent = "app", timeout=None)
```

```python
lat = []
lon = []
for names in rest_loc['Name']:
    location = geolocator.geocode(names)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
print(lat)
```

```python
print(lon)
```

```python
rest_loc['lat'] = lat
rest_loc['lon'] = lon
```
